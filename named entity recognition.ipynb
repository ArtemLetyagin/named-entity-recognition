```python
from flair.data import Sentence
from flair.models import SequenceTagger
```

```python
bloom_data = open('BloombergScraping.txt', 'r', encoding='utf8')
bloom_data = bloom_data.read()
mas_of_data = bloom_data.split('\n\n')
```

```python
mas_of_data = mas_of_data[:1000]
len(mas_of_data)
```

```python
tagger = SequenceTagger.load('ner')
```

```python
mas_ORG = []
mas_LOC = []
for i in range(len(mas_of_data)):
    sentence = Sentence(mas_of_data[i])
    tagger.predict(sentence)
    spans = sentence.get_spans('ner')
    mas_ORG.append([])
    mas_LOC.append([])
    for j in range(len(spans)):
        if(str(spans[j].get_labels()[0]).find('ORG')!=-1):
            mas_ORG[i].append(spans[j].text)
    for j in range(len(spans)):
        if(str(spans[j].get_labels()[0]).find('LOC')!=-1):
            mas_LOC[i].append(spans[j].text)
    print(i)
```

```python
ORG = open('mas_ORG.txt', 'w')
for org in mas_ORG:
    count=0
    if(len(org)>0):
        for i in range(len(org)-2):
            ORG.write(org[i]+', ')
            count+=1
        ORG.write(org[count])
        ORG.write('\n\n')
    else:
        ORG.write('No information'+'\n\n')
LOC = open('mas_LOC.txt', 'w')
for loc in mas_LOC:
    count=0
    if(len(loc)>0):
        for i in range(len(loc)-2):
            LOC.write(loc[i]+', ')
            count+=1
        LOC.write(loc[count])
        LOC.write('\n\n')
    else:
        LOC.write('No information'+'\n\n')
ORG.close()
LOC.close()
```

```python
ORG_file = open('mas_ORG.txt', 'r')
ORG_data = ORG_file.read()
ORG_data = ORG_data.split('\n\n')

LOC_file = open('mas_LOC.txt', 'r')
LOC_data = LOC_file.read()
LOC_data = LOC_data.split('\n\n')
```
